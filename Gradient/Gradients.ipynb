```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def convolution(img, kernel):
    N = img.shape[0]
    M = img.shape[1]

    kN = kernel.shape[0]
    kM = kernel.shape[1]

    kernel = np.array(kernel, dtype=int)
    
    ans = np.zeros(img.shape)
    for i in range(0+kN-1,N):
        for j in range(0+kM-1, M):
            imgKern = np.array(img[i-kN+1:i+1, j-kM+1:j+1], dtype=int)
            imgCheck = imgKern * kernel
            
            ans[i-int(kN/2),j-int(kM/2)] = np.sum(imgCheck)
    return ans
```

```python
def gradient(img, kernel):
    ans = convolution(img, kernel)
    return (ans + 255) // 2 
```

```python
img = cv2.imread('einstein.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

## Simple gradient

```python
kernel = np.array([[-1, 0, 1]])
ans = gradient(img, kernel)
plt.figure()
plt.imshow(ans, cmap='gray')
plt.title('Gradient in X')
plt.show()
```

```python
kernel = np.array([[-1],
                   [0],
                   [1]])
ans = gradient(img, kernel)
plt.figure()
plt.imshow(ans, cmap='gray')
plt.title('Gradient in Y')
plt.show()
```

## Prewitt kernel

```python
kernel = np.array([[-1, 0, 1],
                [-1, 0, 1],
                 [-1, 0, 1]])
ans = gradient(img, kernel)
plt.figure()
plt.imshow(ans, cmap='gray')
plt.title('Gradient in X')
plt.show()
```

```python
kernel = np.array([[-1, -1, -1],
                [0, 0, 0],
                 [1, 1, 1]])
ans = gradient(img, kernel)
plt.figure()
plt.imshow(ans, cmap='gray')
plt.title('Gradient in Y')
plt.show()
```

## Sobel kernel

```python
kernel = np.array([[-1, 0, 1],
                [-2, 0, 2],
                 [-1, 0, 1]])
ans = gradient(img, kernel)
plt.figure()
plt.imshow(ans, cmap='gray')
plt.title('Gradient in X')
plt.show()
```

```python
kernel = np.array([[-1, -2, -1],
                [0, 0, 0],
                 [1, 2, 1]])
ans = gradient(img, kernel)
plt.figure()
plt.imshow(ans, cmap='gray')
plt.title('Gradient in Y')
plt.show()
```

```python
kernel = np.array([[0, 0, 1],
                [0, 0, 0],
                 [-1, 0, 0]])
ans = gradient(img, kernel)
plt.figure()
plt.imshow(ans, cmap='gray')
plt.title('Gradient in 45')
plt.show()
```

# Additional examples

```python
def show_gradients(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    plt.figure(figsize=(15, 6))
    
    plt.subplot(141)
    plt.imshow(img, cmap='gray')
    plt.title('Original')
    plt.xticks([]), plt.yticks([])
    
    kernel = np.array([[-1, 0, 1]])
    ans = gradient(img, kernel)
    plt.subplot(142)
    plt.imshow(ans, cmap='gray')
    plt.title('Gradient in X')
    plt.xticks([]), plt.yticks([])
    
    kernel = np.array([[-1],
                   [0],
                   [1]])
    ans = gradient(img, kernel)
    plt.subplot(143)
    plt.imshow(ans, cmap='gray')
    plt.title('Gradient in Y')
    plt.xticks([]), plt.yticks([])
    
    kernel = np.array([[0, 0, 1],
                [0, 0, 0],
                 [-1, 0, 0]])
    ans = gradient(img, kernel)
    plt.subplot(144)
    plt.imshow(ans, cmap='gray')
    plt.title('Gradient in 45')
    plt.xticks([]), plt.yticks([])
    
    plt.subplots_adjust(bottom=0.1, right=0.8, top=0.9)
    plt.show()
```

```python
show_gradients('rafiki.jpg')
```

```python
show_gradients('test_im_1.png')
```
